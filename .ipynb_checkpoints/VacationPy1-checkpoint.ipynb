```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json

# Import API key
from api_keys import g_key
gmaps.configure(api_key=g_key)
```

### Store Part I results into DataFrame

```python
city_data=pd.read_csv('city_data.csv')
city_data
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
locations = city_data[["Latitude","Longitude"]]

# Fill NaN values and convert to float
rating = city_data["Humidity"].astype(int)


fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=rating,
                                dissipating=False, max_intensity=100,
                                 point_radius=5,opacity = 0.8)


# Add layer


fig.add_layer(heat_layer)

# Display figure
fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
str_test="MHPG+CC Port Hedland, Western Australia"
str_list = str_test.split(',')  # list of words

# first word  v              v last word
str_list[0], str_list[-1]
```

```python
with open('Hotel_info.json') as json_file:
     response= json.load(json_file)
response
Hotel_Name=response["results"][0]["name"]
Hotel_Adress=response["results"][0]["vicinity"]
Hotel_Location=response["results"][0]['geometry']['location']
Hotel_lat=response["results"][0]['geometry']['location']['lat']
Hotel_lng=response["results"][0]['geometry']['location']['lng']
Contry_code=response["results"][0]["plus_code"]['compound_code']
Contry_code_List = Contry_code.split(',')  # list of words
Country_name=str_list[-1]
Country_name
```

```python
target_radius = 50000
target_type = "Hotel"
target_search = "Hotel"
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
Hotel_Names=[]
Hotel_Adresses=[]
Hotel_Locations=[]
Hotel_Lat=[]
Hotel_Lng=[]
Best_cities_List=Best_cities['City']
City_names=[]
Country=[]

# set up a parameters dictionary
i=0
for City in Best_cities_List:
    response=[]
    target_coordinates=str(Best_cities.iloc[i,2])+","+str(Best_cities.iloc[i,3])
    params = {
        "location": target_coordinates,
        "radius": target_radius,
        "type": target_type,
        "key": g_key,
        "keyword": target_search
        }
    try:
        response = requests.get(base_url, params=params).json()
        Hotel_Name=response["results"][0]["name"]
        Hotel_Adress=response["results"][0]["vicinity"]
        Hotel_Location=response["results"][0]['geometry']['location']
        Hotel_lat=response["results"][0]['geometry']['location']['lat']
        Hotel_lng=response["results"][0]['geometry']['location']['lng']
        Contry_code=response["results"][0]["plus_code"]['compound_code']
        
        Contry_code_List = Contry_code.split(',')  # list of words
        Country_name=Contry_code_List[-1]
        
        
        Country.append(Country_name)
        City_names.append(City)
        Hotel_Adresses.append(Hotel_Adress)
        Hotel_Locations.append(Hotel_Location)
        Hotel_Names.append(Hotel_Name)
        Hotel_Lat.append(Hotel_lat)
        Hotel_Lng.append(Hotel_lng)

        
        print (f'We found The best hotel in {City}')
    except:
        print(f'Could not find any {target_search} in {target_radius} meters of {City}')
    #if i==2: break
    i=i+1
# base url

# run a request using our params dictionary
Country
```

```python
Best_Hotel_data = pd.DataFrame({
                                'Country':Country,
                                
                                'City': City_names,
                               'Name': Hotel_Names,
                               'Latitude': Hotel_Lat,
                               'Longitude': Hotel_Lng,
                                'Adress': Hotel_Adresses,
                                'Hotel_Locations': Hotel_Locations,
                            
                                })
#Test=city_data.loc[city_data['City']==Best_Hotel_data['City']]
#Best_Hotel_data.to_csv('Best_Hotel_Data.csv')
Best_Hotel_data

```
